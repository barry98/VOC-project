```python
test = []

for x in ast.literal_eval(df.namen[0]):
    if x['tussenvoegsel'] != None:
        test.append(x['voornaam'] + " " + x['tussenvoegsel'] + " " + x['achternaam'])
    else:
        test.append(x['voornaam'] + " " + x['achternaam'])
```

```python
for ent in doc.ents:
    if ent.text in test:
        print(ent.text, ent.start_char, ent.end_char, ent.label_)
```
