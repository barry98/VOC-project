```python
import pandas as pd
import numpy as np
from wordcloud import WordCloud, STOPWORDS 
import matplotlib.pyplot as plt 
```

```python
voc = pd.read_csv('../vocclustered.csv', sep='	')
```

### Namen

```python
# Standardized Names
wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='white', 
                min_font_size = 10).generate_from_frequencies(voc.fullNameNormalized.value_counts())

plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show() 
```

```python
# Original Names
wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='white', 
                min_font_size = 10).generate_from_frequencies(voc.fullNameOriginal.value_counts())

plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show() 
```

```python
objects = [x for x in voc.fullNameNormalized.value_counts().keys()[0:50]]
y_pos = np.arange(len(objects))
performance = [x for x in voc.fullNameNormalized.value_counts()[0:50]]

plt.figure(figsize = (20, 8)) 
plt.bar(y_pos, performance, align='center', alpha=0.5)
plt.xticks(y_pos, objects, rotation=80)
plt.ylabel('Entries')
plt.title('50 most common names (normalized)')

plt.show()
```

```python
objects = [x for x in voc.fullNameOriginal.value_counts().keys()[0:50]]
y_pos = np.arange(len(objects))
performance = [x for x in voc.fullNameOriginal.value_counts()[0:50]]

plt.figure(figsize = (20, 8)) 
plt.bar(y_pos, performance, align='center', alpha=0.5)
plt.xticks(y_pos, objects, rotation=80)
plt.ylabel('Entries')
plt.title('50 most common names (original)')

plt.show()
```

### Rang

```python
objects = [x for x in voc['rank'].value_counts().keys()[0:10]]
y_pos = np.arange(len(objects))
performance = [x for x in voc['rank'].value_counts()[0:10]]

fig, ax = plt.subplots()

ax.barh(y_pos, performance, align='center')
ax.set_yticks(y_pos)
ax.set_yticklabels(objects)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Entries')
ax.set_title('10 most common ranks')

plt.show()
```

```python
objects = [x for x in voc['rank'].value_counts().keys()[3:13]]
y_pos = np.arange(len(objects))
performance = [x for x in voc['rank'].value_counts()[3:13]]

fig, ax = plt.subplots()

ax.barh(y_pos, performance, align='center')
ax.set_yticks(y_pos)
ax.set_yticklabels(objects)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Entries')
ax.set_title('10 most common ranks outside of most common 3')

plt.show()
```

```python
objects = [x for x in voc['rank'].value_counts().keys()[-10:]]
y_pos = np.arange(len(objects))
performance = [x for x in voc['rank'].value_counts()[-10:]]

fig, ax = plt.subplots()

ax.barh(y_pos, performance, align='center')
ax.set_yticks(y_pos)
ax.set_yticklabels(objects)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Entries')
ax.set_title('10 least common ranks')

plt.show()
```
